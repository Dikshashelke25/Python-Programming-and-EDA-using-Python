### Task 22 - Groupby and Crosstab:

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df_loan = pd.read_csv('loandata.csv')
```

```python
df_loan
```

#### 1.Display the first 5 rows of the DataFrame using head().


```python
df_loan.head(5)
```

#### a.Display the unique values in the Gender column.


```python
df_loan['Gender'].unique()
```

#### 2.Group the DataFrame by the Gender column and calculate the mean of ApplicantIncome for each gender using groupby().


```python
df_loan.groupby('Gender')['ApplicantIncome'].mean()
```

#### a.Calculate the minimum LoanAmount for each Education category using groupby().


```python
df_loan.groupby('Education')['LoanAmount'].min()
```

#### 3.Group the DataFrame by both Gender and Married columns, and calculate the maximum LoanAmount and mean ApplicantIncome using agg().


```python
df_loan.groupby(['Gender','Married'])['LoanAmount'].agg(['max'])
```

```python
df_loan.groupby(['Gender','Married'])['ApplicantIncome'].agg(['mean'])
```

#### a.Calculate the count, variance, and standard deviation of LoanAmount for each Property_Area using .agg([np.var, np.max, np.std, np.min,np.max, np.count_nonzero]).


```python
df_loan.groupby(['Property_Area'])['LoanAmount'].agg([np.var, np.std, np.min,np.max, np.count_nonzero])
```

#### 4.Sort the DataFrame by LoanAmount in descending order


```python
df_loan.sort_values('LoanAmount', ascending = False)
```

#### a.find the number of applicants with different Loan_Status values.


```python
df_loan['Loan_Status'].value_counts()
```

#### 5.Create a cross-tabulation of Gender and Loan_Status to see the count of loan approvals based on gender using crosstab().


```python
pd.crosstab(df_loan.Gender,df_loan.Loan_Status)
```

#### 6.Create a pivot table that shows the average LoanAmount for each combination of Education and Property_Area.


```python
df_loan.pivot_table(values='LoanAmount',index='Education',columns='Property_Area',aggfunc='mean')
```

#### a.Group the data by Credit_History and Loan_Status and calculate the mean, max, and min of ApplicantIncome using groupby() and .agg().


```python
df_loan.groupby(['Credit_History','Loan_Status'])['ApplicantIncome'].agg([np.mean, np.max, np.min])
```

#### 7.Sort the DataFrame by ApplicantIncome in ascending order, then by LoanAmount in descending order using sort_values().


```python
df_loan.sort_values(['ApplicantIncome','LoanAmount'], ascending = [True, False])
```

#### a.Create a pivot table to calculate the sum of LoanAmount for each Self_Employed status, grouped by Gender.


```python
df_loan.pivot_table(values='LoanAmount',index='Gender',columns='Self_Employed',aggfunc='sum')
```
