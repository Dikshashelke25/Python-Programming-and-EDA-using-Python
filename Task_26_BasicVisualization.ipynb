1. You have a DataFrame monthly_sales with columns 'Month' and 'Sales'. Create a line plot to visualize the trend of monthly sales over time.
   
   data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
        	   'Sales': [100, 120, 90, 110, 130]}


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Creating the DataFrame
data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'], 'Sales': [100, 120, 90, 110, 130]}
monthly_sales = pd.DataFrame(data)
monthly_sales
```

```python
# Plotting the line plot
plt.figure(figsize=(8, 5))
sns.lineplot(x='Month', y='Sales', data=monthly_sales, marker='o')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Trend')
plt.show()
```

8. Calculate Range, Variance and Standard deviation of incomes.

    Write a Python function that takes this dataset as input, performs the above analysis, and returns the results


```python
def income_statistics(employee_salaries):
    
    income_range = np.max(employee_salaries['Salary']) - np.min(employee_salaries['Salary'])
    income_variance = np.var(employee_salaries['Salary']) 
    income_std_dev = np.std(employee_salaries['Salary'])
    
    return {
        'Range': income_range,
        'Variance': income_variance,
        'Standard Deviation': income_std_dev
    }
result = income_statistics(employee_salaries)
print(result)
```

9. You are given a dataset containing information about the monthly income of a few individuals. Perform the following analysis using Pandas:

          data = {'Monthly Income': [5000, 6000, 7000, 5500, 6500, 7500, 8000, 8500, 9000, 9500]}

       Calculate the mean, median, and mode of the monthly income.


```python
# Creating the DataFrame
data = {'Monthly Income': [5000, 6000, 7000, 5500, 6500, 7500, 8000, 8500, 9000, 9500]}
income_df = pd.DataFrame(data)
```

```python
mean_income = income_df['Monthly Income'].mean() 
median_income = income_df['Monthly Income'].median()  
mode_income = income_df['Monthly Income'].mode() 

print(f"Mean Monthly Income: {mean_income}")
print(f"Median Monthly Income: {median_income}")
print(f"Mode Monthly Income: {mode_income.tolist()}")
```
