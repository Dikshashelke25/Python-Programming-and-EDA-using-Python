#### 6.What is the range of values for the ‘SepalLengthCm’ and ‘PetalWidthCm’?


```python

sepal_length_range = (df['SepalLengthCm'].min(), df['SepalLengthCm'].max())
petal_width_range = (df['PetalWidthCm'].min(), df['PetalWidthCm'].max())

print(f"Range of SepalLengthCm: {sepal_length_range}")
print(f"Range of PetalWidthCm: {petal_width_range}")
```

#### 7.What is the mean and median of the ‘PetalLengthCm’ column?


```python
PL_mean = df['PetalLengthCm'].mean()
PL_median = df['PetalLengthCm'].median()

print(f"Mean of PetalLengthCm: {PL_mean}")
print(f"Median of PetalLengthCm: {PL_median}")
```

#### 8.What does the df.describe().T function do, and how can it help in performing exploratory data analysis on the Iris dataset? Interpret the output.


```python
df.describe()
```

The .describe() method in pandas provides a summary of statistical data for each numerical column in a DataFrame. Using .T transposes the output so that rows become columns and vice versa.

#### 11.How would you filter the dataset to include only rows where 'SepalLengthCm' is greater than 5.0 and 'PetalWidthCm' is less than 1.5 using the & operator?

#### 12.Using loc and & to Select Rows Labeled 'Versicolor' where 'SepalLengthCm' is Between 5.5 and 6.5


```python
df.loc[(df.Species == 'Iris-versicolor') & (df.SepalLengthCm.between(5.5, 6.5))]
```

#### 13.Using loc and | to Filter Rows Labeled 'Virginica' or 'Setosa' with 'Petal'WidthCm' Greater Than 2.0


```python
df.loc[((df.Species == 'Iris-virginica') | (df.Species == 'Iris-setosa')) & (df.PetalWidthCm > 2)]
```
